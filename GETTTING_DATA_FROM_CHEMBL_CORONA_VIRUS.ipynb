**DATA PRE-PROCESSING **

DRUGS HAVING MORE THAN 1000NM WILL BE ACTIVE

1.   DRUGS HAVING MORE THAN 10000NM WILL BE ACTIVE

2.   DRUGS HAVING LESS THAN 1000NM WILL BE IN ACTIVE.



1. DRUGS HAVING BETWEEEN 1000NM AND 10,0000 WILL BE INTERMIDATE 

2.   List item






```python
bioactivity_class= [ ]
for i in df2.standard_value :
  if float (i) >= 1000 :
    bioactivity_class.append ("inactive")
  
  elif  float (i) <= 1000:
    bioactivity_class.append ("active")
  else:
    bioactivity_class.append ("intermidate ")
    
```

**	molecule_chembl_id to a list **

```python
mol_cid = []
for i in df2.	molecule_chembl_id:
  mol_cid.append(i)
```

```python
selection = ["molecule_chembl_id" ,  "canonical_smiles" , "standard_value" ]
df3= df2[selection]
```

```python
 data = pd.concat([df3,pd.Series(bioactivity_class)],axis= 1)
```

```python
data.columns
```

```python
 data=data.rename (columns= {"molecule_chembl_id" : "Molecule_chembl_id" , "canonical_smiles" : "Canonical_Smiles", "standard_value" : "Standard_Value "  , 0: "Bioactivity_class" })
```

```python
df3=data
```

```python
df3.columns
```

```python
df3=df3[['Molecule_chembl_id', 'Canonical_Smiles', 'Standard_Value ',
       #'bioactivity_class',
       'Bioactivity_class']].copy()
```
